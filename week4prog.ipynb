# Intro
## Распределение Максвелла:
$$f(x)=\sqrt{2/\pi}*x^2*e^{-x^2/2}$$
$$ M(f)=\int_0^{\infty} \sqrt{2/\pi}*x^2*e^{-x^2/2} x dx\approx 1.59 $$
$$ D(f)=M(f^2)-M^2(f)\approx0.45 $$
## Для того, чтобы определить параметры нормального распределения, аппроксимирующего график для различных n, применим ЦПТ:
$$ \overline{X_n} \sim N(M(f),\frac{D(f)}{n}) $$
$$n=5: D(f)=0.09 $$
$$n=10: D(f)=0.045$$
$$n=50: D(f)=0.009$$

```python
import pandas as pd
import pylab
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline




#N=5


# Задаю распределение
max_rv = sts.maxwell(0)
sum=0.0

spisok=[]
# Заполняю список среднего значения по выборкам
for j in range(1000):
    sample = max_rv.rvs(5)
  
    for i in sample:
        sum=sum+i
    sum=sum/5.0
    spisok.append(sum)
    sum=0.0


pylab.figure(1)
# Задаю нормальное распределение и строю графики
norm_rv = sts.norm(1.59, 0.3)
sample = norm_rv.rvs(1000)
x = np.linspace(0,3,1000)
pdf = norm_rv.pdf(x)
pylab.plt.plot(x, pdf, label='theoretical pdf(n=5)', alpha=0.5)
pylab.plt.legend()

pylab.plt.ylabel('$f(x)$')
pylab.plt.xlabel('$x$')
pylab.plt.hist(spisok, bins=10, normed=True)


pylab.plt.ylabel('fraction of samples')
pylab.plt.xlabel('$x$')




#N=10
max_rv = sts.maxwell(0)
sum=0.0

spisok=[]
for j in range(1000):
    sample = max_rv.rvs(10)
  
    for i in sample:
        sum=sum+i
    sum=sum/10.0
    spisok.append(sum)
    sum=0.0




norm_rv = sts.norm(1.59, 0.21)
sample = norm_rv.rvs(1000)
x = np.linspace(0,3,1000)
pdf = norm_rv.pdf(x)


pylab.figure (2)
pylab.plot (x, pdf, label='theoretical pdf(n=10)', alpha=0.5)
pylab.legend()



pylab.plt.ylabel('$f(x)$')
pylab.plt.xlabel('$x$')
pylab.plt.hist(spisok, bins=10, normed=True)


pylab.plt.ylabel('fraction of samples')
pylab.plt.xlabel('$x$')









#N=50

max_rv = sts.maxwell(0)
sum=0.0

spisok=[]
for j in range(1000):
    sample = max_rv.rvs(50)
  
    for i in sample:
        sum=sum+i
    sum=sum/50.0
    spisok.append(sum)
    sum=0.0



norm_rv = sts.norm(1.59, 0.1)
sample = norm_rv.rvs(1000)
x = np.linspace(0,3,1000)
pdf = norm_rv.pdf(x)


pylab.figure (3)
pylab.plot (x, pdf, label='theoretical pdf(n=50)', alpha=0.5)
pylab.legend()



pylab.plt.ylabel('$f(x)$')
pylab.plt.xlabel('$x$')
pylab.plt.hist(spisok, bins=10, normed=True)


pylab.plt.ylabel('fraction of samples')
pylab.plt.xlabel('$x$')

```

# Вывод
Как мы видим: чем больше объем выборки(n), тем точнее гистограмма ее среднего описывается нормальным распределением
